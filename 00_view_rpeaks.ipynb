```python
import wfdb as wf
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
from ecgdetectors.ecgdetectors import Detectors
```

```python
# from glob import glob

def get_records():
    """ Get paths for data in data/mitdb/ directory """
    # There are 3 files for each record
    # *.atr is one of them
    paths = glob('data/mitdb/*.atr')

    # Get rid of the extension
    paths = [path[:-4] for path in paths]
    paths.sort()

    return paths
```

# Load File

```python
# Change the values below to change your data selection.

FILE          = 7
CHANNEL       = 0
SAMPLE_START  = 1
SAMPLE_SIZE   = 5000
FS            = 360

# Get available data files.
dbfiles = get_records()
print('Total files: ', len(dbfiles))

# Pick a file.
datfile = dbfiles[FILE]
print('Loading file:', datfile)

# Read the file.
record = wf.rdsamp(datfile)
annotation = wf.rdann(datfile, 'atr')

# Get data and annotations for the samples selected below.
sample_end = SAMPLE_START + SAMPLE_SIZE
channel = record[0][SAMPLE_START:sample_end, CHANNEL]

# Info about the data.
print('Sampling frequency:', record[1].get('fs'))
print('Data shape:', record[0].shape)
print('Annotations:', len(annotation.num))

signal_pad_samples = 10
signal_pad = np.zeros(signal_pad_samples) # pad one sec to detect initial peaks properly

# Detect R-Peaks
detectors = Detectors(FS)
# rpeaks = detectors.christov_detector(channel)
# rpeaks = np.array(rpeaks)
detectors = {
            'pan_tompkins_detector':[detectors.pan_tompkins_detector, []],
            'hamilton_detector':[detectors.hamilton_detector, []],
            'christov_detector':[detectors.christov_detector, []],
            'engzee_detector':[detectors.engzee_detector, []],
            'swt_detector':[detectors.swt_detector, []],
            'two_average_detector':[detectors.two_average_detector, []],
            }

for kd in detectors.keys():
    vd = detectors[kd]
    r_peaks = np.array(vd[0](np.hstack((signal_pad, channel)))) - signal_pad_samples

    plt.figure(figsize=(16,2))
    plt.title(kd)
    plt.plot(channel, color='tab:blue',linewidth=0.8)
    plt.xticks(r_peaks)
    plt.grid(axis='x')
    plt.show()
```
